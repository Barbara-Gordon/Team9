```python
import csv
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
charging_stations_df = pd.read_csv("charging_stations.csv")
charging_stations_df.head()
```

```python
charging_stations_df = charging_stations_df[charging_stations_df['Fuel Type Code'] == "ELEC"]
```

```python
charging_stations_df = charging_stations_df.groupby(['City', 'State', 'ZIP'], as_index=False).agg({'Latitude': 'mean',
                                                 'Longitude': 'mean', 'Station Name': 'count'})
charging_stations_df.head()
```

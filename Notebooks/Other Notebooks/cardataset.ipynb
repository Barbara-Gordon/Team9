```python
#Importing Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading Data for cars
df_car_meta=pd.read_csv("Datasets/cardataset.zip")

# Renaming column names by adding '_' 
df_car_meta.columns = [c.replace(' ', '_') for c in df_car_meta.columns]

#removed this row --> was wrong data and the correct MPG is 34
df_car_meta.drop(df_car_meta[df_car_meta.highway_MPG == 354].index, inplace=True)

#Categorizing fuel type broadly into premium,diesel,flex-fuel,electric & natural gas
df_car_meta.replace(to_replace='premium unleaded (required)', value='Premium', inplace=True)
df_car_meta.replace(to_replace='premium unleaded (recommended)', value='Premium', inplace=True)
df_car_meta.replace(to_replace='flex-fuel (unleaded/E85)', value='Flex-Fuel', inplace=True)
df_car_meta.replace(to_replace='flex-fuel (premium unleaded recommended/E85)', value='Flex-Fuel', inplace=True)
df_car_meta.replace(to_replace='flex-fuel (premium unleaded required/E85)', value='Flex-Fuel', inplace=True)
df_car_meta.replace(to_replace='flex-fuel (unleaded/natural gas)', value='Flex-Fuel', inplace=True)
df_car_meta.replace(to_replace='regular unleaded', value='Regular', inplace=True)
df_car_meta.replace(to_replace='diesel', value='Diesel', inplace=True)
df_car_meta.replace(to_replace='electric', value='Electric', inplace=True)
df_car_meta.replace(to_replace='natural gas', value='Natural Gas', inplace=True)
```
